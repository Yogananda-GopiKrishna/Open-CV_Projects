### Create a function which help to find coordinate of any  pixel   its color

```python

def mouse_event(event,x,y,flags,param):
    print("x",x)
    print("y",y)
    print("flag",flags)
    print("param",param)
    font = cv2.FONT_HERSHEY_COMPLEX
    if event == cv2.EVENT_LBUTTONDOWN:
        print(x," ",y)
        coord = ". "+str(x)+", "+str(y)
        cv2.putText(img,coord,(x,y),font,1,(0,0,255),2,)
        cv2.imshow("image",img)
    if event == cv2.EVENT_RBUTTONDBLCLK:
        b = img[y,x,0]   #for blue channel is 0
        g = img[y,x,1]   #for green channel
        r = img[y,x,2]   #for red channel

        color_bgr = ". "+str(b)+", "+str(g)+", "+str(r)
        cv2.putText(img,color_bgr,(x,y),font,1,(0,255,0),2)
        cv2.imshow('image',img)

cv2.namedWindow(winname='res')

img = cv2.imread("thor.jpg")
cv2.setMouseCallback("res",mouse_event)


while True:
    cv2.imshow('res',img)
    if cv2.waitKey(1) & 0xFF ==27 :  #esc
        break

cv2.destroyAllWindows()
```
