```python
import cv2
import numpy as np

#camera = "http://192.168.1.4:8080//video"

#read camera
cap = cv2.VideoCapture(0,cv2.CAP_DSHOW)
def nothing(x):
    pass

#mobile camera access
#cap.open(camera)
#print("check",cap.isOpened())

#window name
cv2.namedWindow("color adj",cv2.WINDOW_NORMAL)
cv2.resizeWindow("color adj",(300,300))

cv2.createTrackbar("thres","color adj",0,255,nothing)

#color detection track
cv2.createTrackbar("Lower_H","color adj",0,255,nothing)
cv2.createTrackbar("Lower_S","color adj",0,255,nothing)
cv2.createTrackbar("Lower_V","color adj",0,255,nothing)
cv2.createTrackbar("Higher_H","color adj",255,255,nothing)
cv2.createTrackbar("Higher_S","color adj",255,255,nothing)
cv2.createTrackbar("Higher_V","color adj",255,255,nothing)


while cap.isOpened():
    ret,frame = cap.read()
    frame = cv2.flip(frame,2)
    frame = cv2.resize(frame,(512,224))
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

    #Detecting  hand
    Lower_H = cv2.getTrackbarPos("Lower_H","color adj")
    Lower_S = cv2.getTrackbarPos("Lower_S","color adj")
    Lower_V = cv2.getTrackbarPos("Lower_V","color adj")
    Higher_H = cv2.getTrackbarPos("Higher_H","color adj")
    Higher_S = cv2.getTrackbarPos("Higher_S","color adj")
    Higher_V = cv2.getTrackbarPos("Higher_V","color adj")

    upper_bound = np.array([Higher_H,Higher_S,Higher_V])
    lower_bound = np.array([Lower_H,Lower_S,Lower_V])

    #creating mask
    mask = cv2.inRange(hsv,lower_bound,upper_bound)
        
    #filter mask with image
    filter = cv2.bitwise_and(frame,frame,mask=mask)
        
    #background black and foreground image
    mask1 = cv2.bitwise_not(mask)
        
    mask_thres = cv2.getTrackbarPos("thres","color adj") #get track bar value
    #applying some simple threshold value
    ret,thresh  = cv2.threshold(mask1,mask_thres,255,cv2.THRESH_BINARY)
    #dilate = cv2.dilate(thresh,(1,1),iterations=2)
    #e = cv2.erode(thresh,(1,1),iterations=2) 
    canon = cv2.Canny(thresh,2,2)
    
    #find contours 
    cnts,hier = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

    #cnts : a list of contours and each consists an array of x,y coordinates
    #hier : contains image information

    #print("contours",cnts)

    #frame = cv2.drawContours(frame,cnts,-1,(0,255,0),-1)

    for c in cnts:
        area = cv2.contourArea(c)
        eps = 0.00001* cv2.arcLength(c,closed=True)
        data = cv2.approxPolyDP(c,eps,True)

        hull = cv2.convexHull(data)

        cv2.drawContours(frame,[c],-1,(50,100,200),3)
        cv2.drawContours(frame,[hull],-1,(0,255,0),1)
         

    cv2.imshow("Threshold",thresh)
    cv2.imshow("mask",mask)
    cv2.imshow("filter",filter)
    cv2.imshow("Result",frame)

    if cv2.waitKey(25) & 0xFF == 27:
            break

cap.release()
cv2.destroyAllWindows()













```
